```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
# image is always read as bgr channel's by default in cv
# to view the image in normally i.e rgb format we neet convert the color space
```

```python
img_bgr = cv.imread('../DATA/00-puppy.jpg')
```

```python
img_rgb = cv.cvtColor(img_bgr, cv.COLOR_BGR2RGB)
```

```python
plt.imshow(img_rgb)
```

```python
# Reading Image as a GrayScale (Black & White)
# Method 1 Opening Directly as GrayScale 
# Method 2 Opening Directly as GrayScale by 0 channels
# Method 3 Converting Img to GrayScale
```

```python
img_gs1 = cv.imread('../DATA/00-puppy.jpg', cv.IMREAD_GRAYSCALE)
```

```python
plt.imshow(img_gs1, cmap='gray')
```

```python
img_gs2 = cv.imread('../DATA/00-puppy.jpg', 0)
```

```python
plt.imshow(img_gs2, cmap='gray')
```

```python
img_gs3 = cv.cvtColor(img_bgr, cv.COLOR_BGR2GRAY)
```

```python
plt.imshow(img_gs3, cmap='gray')
```
